```python
from random import random
import matplotlib.pyplot as plt
import numpy as np

N=pow(10, 5)
b = 5.0 # x_max
a = 0.0 # x_min
E_lista = [] # lista de aleatórios
fE_lista = []
fE2_lista = [] #f(E)²

def f(E):
    return 1/(1 + E**2)

def Integracao(fE_lista): # método de integração
    return ((b-a)/N)*sum(fE_lista)

for i in range(N):
    E_random = b*random()
    fE = f(E_random) # f(E)
    fE2 = fE**2 # f(E)²
    E_lista.append(E_random)
    fE_lista.append(fE)
    fE2_lista.append(fE2)
    
I1 = Integracao(fE_lista)
    
#plt.scatter(x_lista, y_lista)
print('integral calculada:', I1, '\nValor Real:', np.arctan(b)-np.arctan(a))
```

### __b.__

```python
def var_g(): # variância
    return ( (1/N)*((b-a)**2)*sum(fE2_lista) ) - ((1/N)*(b-a)*sum(fE_lista) )**2
def std_g(): # desvio padrão
    return np.sqrt(var_g())
def err_g(): # erro-padrão
    return std_g()/np.sqrt(N)
print('var:',var_g(), '\nstd', std_g(), '\nerro', err_g())
```
