<b>alínea b)</b>

A distribuição triangular tem como PDF:

\begin{align*}
        \begin{cases}
            0 & \text{para } E < a, \\
            \frac{2(E-a)}{(b-a)(c-a)} & \text{para } a \le E < c, \\[4pt]
            \frac{2}{b-a}             & \text{para } E = c, \\[4pt]
            \frac{2(b-E)}{(b-a)(b-c)} & \text{para } c < E \le b, \\[4pt]
            0 & \text{para } b < E.
        \end{cases}  
\end{align*}

O CDF é calculável integrando a função acima:

\begin{align*}
        \begin{cases}
            \int_0^a 0dE=0 & \text{para } 0< E < a, \\
            \int_a^E\frac{2(E-a)}{(b-a)(c-a)}dE & = \frac{2}{(b-a)(c-a)}\int_a^E(E-a)dE & = \frac{(E-a)^2}{(b-a)(c-a)}& \text{para } a \le E \le c, \\[4pt]
            \frac{(c-a)^2}{(b-a)(c-a)}+\int_c^E\frac{2(b-E)}{(b-a)(b-c)}dE & = \frac{c-a}{b-a}+\frac{2}{(b-a)(b-c)}\int_c^E(b-E)dE & = \frac{c-a}{b-a}+\frac{b-c}{b-a}-\frac{(b-E)^2}{(b-a)(b-c)}&=1-\frac{(b-E)^2}{(b-a)(b-c)} & \text{para } c < E \le b, \\[4pt]
            1 & \text{para } b < E.
        \end{cases}  
\end{align*}

A transformada inversa também pode ser calculada:
\begin{align*}
        \begin{cases}
           F(E)=U=\frac{(f(E)-a)^2}{(b-a)(c-a)}& \Leftrightarrow f(E)=a+\sqrt{U(b-a)(c-a)}& \text{para } U \le F(c)=\frac{c-a}{b-a} \\[4pt]
           F(E)=U=1-\frac{(b-f(E))^2}{(b-a)(b-c)} & \Leftrightarrow f(E)=b-\sqrt{U(b-a)(b-c)} & \text{para } U > F(c)=\frac{c-a}{b-a} \\[4pt]
        \end{cases}  
\end{align*}

Agora é só amostrar para $U \sim Unif(0,1)$ e daí retirar a distribuição triangular:


```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
from random import random
from scipy import special
amostra=[]
n=20
a=1
b=9
c=3
y2=[]
x=[]
x2=np.arange(0,20,0.01)
for x2val in x2:
    if x2val>0 and x2val<a:
        y=0.000
        y2.append(y)
    elif x2val>=a and x2val<c:
        y2.append(2*(x2val-a)/((b-a)*(c-a)))
    elif x2val==c:
        y2.append(2/(b-a))
    elif x2val>c and x2val<=b:
        y2.append(2*(b-x2val)/((b-a)*(b-c)))
    elif x2val>b:
        y2.append(0)
y2.append(0)
for i in range(0,n):
    r=random()
    if r<(c-a)/(b-a):
        y=a+np.sqrt(r*(b-a)*(c-a))
        x.append(i)
        amostra.append(y)
    else:
        y=b-np.sqrt((1-r)*(b-a)*(b-c))
        x.append(i)
        amostra.append(y)
dados={'Energy(MeV)':amostra}
bins=np.arange(0,10,.01)
df=pd.DataFrame(data=dados).hist(bins=bins,density=1)
plt.plot(x2,y2,linewidth=3,label='função triangular')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.savefig('gaussiana_trans')
df2=pd.DataFrame(data=dados)
df2
```

<b>Exercício 2.</b>

O integral pode ser calculado por Monte Carlo:

$ I=\lim_{N\to\infty} \frac{b-a}{N}\sum_i f(\xi_i)$, com $f(x)=\frac{1}{1+x^2}$

A variância é dada por:

$var(f(x))=\left(b-a\right)^2 \left[ \frac{1}{N}\sum_i f^2(\xi_i)-\left(\frac{1}{N}\sum_i f(\xi_i)\right)^2 \right]$

O desvio-padrão é:

$\sigma=\sqrt{var(f(x)}$

O erro-padrão é:

$\epsilon=\frac{\sigma}{\sqrt{N}}$

A FOM é:

$FOM=\frac{1}{\epsilon^2 T}$

```python
%matplotlib inline
from random import seed
from random import random
import os
import os
import math
from math import pi
from math import sqrt
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import time

x2=[0]
y2=[0]
y3=[0]
res=[1.373]
dev=[0]
start=time.time()
for N in range (1,10000,1):
    x=0.00000000
    u=0.00000000
    a=1.373
    total=0.00000
    for i in range(N):
        r=5.0*random()
        x = x + 1/(1+r**2)
        u = u + (1/(1+r**2))**2
    os.system('clear')
    float(total)
    total = (5.0/N)*x 
    var=(25/N)*u-(25/N**2)*x**2
    stdev=sqrt(var)
    stderr=1/np.sqrt(N)*stdev
    x2.append(N)
    y2.append(total)
    y3.append(stderr)
    res.append(a)
end=time.time()
time=end-start
fom=1/(time*stderr**2)
print('Valor do integral:',total)
print('Variância:',var)
print('Desvio padrão:',stdev)
print('Erro padrão:',stderr)
print('FOM:',fom)


plt.plot(x2, y2,'-o',color='blue',linewidth=3)
plt.plot(x2, res,color='orange',linewidth=3)

plt.xlabel('N')
plt.ylabel('Integral')
```
