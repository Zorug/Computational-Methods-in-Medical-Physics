# __2. Números aleatórios e sequências de números aleatórios__

## __2.1.1 Método de Von Neumann (método de quadrar e tirar o meio)__

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
seed = int ( input ( " Por favor introduza um numero com n digitos ( par ) :\n [####]" ) )
number = seed
strnumber =( str ( number ) )
n = sum ( c.isdigit () for c in strnumber )
already_seen = set ()
counter = 0
numerator =10**n
numero =[ seed / numerator ]
contagem = [0]
```

```python
while number not in already_seen :
    counter += 1
    already_seen.add( number )
    number2 = number * number
    strnumber = str ( number2 ) [ int ( n /2) : int (2* n - n /2) ].zfill(2* n)
    number = int( strnumber )
    print(f"#{ counter }: { number }")
    numero.append( number / numerator )
    contagem.append( counter )
```

```python
print( n )
print( int ( n /2) )
print( int (2* n - n /2) )
print ( f" Comecamos com a semente { seed } e "
       f" comecou a haver repeticao do numero { number } "
       f" com { counter } iteracoes . { os.linesep } Em seguida a visualizacao da sequencia de numeros produzida normalizada a 1 " )
plt.plot( contagem , numero , '-o')
```

## __3.1 Métodos de rejeição históricos. Cálculo de π__

```python
from random import seed
from random import random
import os
import math
from math import pi
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# figure , chart = plt.subplots (1 , 1 , figsize = (12 ,6) )
# integrar usando numeros aleatorios x e y de 0 a 1
err =[]
pi3 =[]
fig = plt.figure(figsize=(6 ,6))
ax = fig.add_subplot(111)
plt.ion()
fig.show()
fig.canvas.draw()
tamanho = 15 # 30
for N in range (1, tamanho, 1):
    area = 0
    pi2 = 0
    for i in range(N):
        x = -1+2*random() # x aleatório
        y = -1+2*random() # y aleatório
        #a = math.pi
        r = x**2+y**2 # r aleatório
        
        if r <=1:
            area = area + 1
            ax.scatter(x, y, color = 'g', s=30)
            fig.canvas.draw()
        else:
            ax.scatter(x, y, color = 'r', s=30)
            fig.canvas.draw()
            
    pi2 =4/ N * area
    pi3.append( pi2 )
    err.append( np.sqrt ( np.var ( pi3 ) ) )
print ('valor de pi = ' , math.pi )
print ('resultado = ' , pi2 )
print ('erro = ' , np.sqrt ( np.var(pi3)))
```

## __3.4 Amostragens de distribuições. Método da transformada inversa__

```python
amostra =[]
x =[]
x2 = np.arange (0 ,100 ,0.1)
y2 = np.exp(-x2) # f(x)
n =50000
for i in range (1, n) :
    r = random()
    y = -np.log(r)
    x.append(i)
    amostra.append(y)
plt.plot(x , amostra , '-o')
dados ={ 'y ': amostra }
bins = np.arange(0 ,10 ,.01)
df = pd.DataFrame( data = dados ).hist( bins = bins , density =1)
plt.plot(x2, y2, linewidth=3, label='funcao exponencial')
plt.xlim(0, 5)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
from scipy import special

amostra =[]
x =[]
n = 40000
mu =0
sigma =1
x2 = np.arange (-5, 5, 0.01)
y2 =1/( np.sqrt (2* np . pi ) * sigma ) * np.exp ( -0.5*((( x2 - mu ) / sigma ) ) **2)
for i in range (1 , n) :
    r = random ()
    y = mu + sigma * math.sqrt (2) * special.erfinv (2* r -1) # Inverse of the error function
    x.append ( i )
    amostra.append( y )
dados ={ 'y ': amostra }
bins = np.arange (-5, 5, .01)
df = pd.DataFrame ( data = dados ).hist ( bins = bins , density =1)
plt.plot ( x2 , y2 , linewidth =3 , label = ' gaussiana ')
plt.legend ()
plt.xlabel ('x')
plt.ylabel ('f(x)')
```

## __3.5 Amostragens de distribuições. Método da transformada de Box-Müller__

```python
amostra1 =[]
amostra2 =[]
x =[]
mu =0
sigma =1
x2 = np.arange ( -5 ,5 ,0.01)
y2 =1/( np.sqrt(2*np.pi) * sigma) * np.exp(-0.5*(((x2 - mu) / sigma)) **2)
n =200000
for i in range (0, n):
    r1 = -1+2* random() # [-1 e 1]
    r2 = -1+2* random()
    s = r1**2+ r2**2
    if s !=0 and s <=1:
        s1 = -2* math.log(s)/s
        z1 = r1 * (s1**0.5) # X
        z2 = r2 * (s1**0.5) # Y
        amostra1.append(z1)
        amostra2.append(z2)
        x.append( i )
# plt.plot (x , amostra1 , '-o')
dados ={'y':amostra1}
bins = np.arange( -5 ,5 ,.01)
df = pd.DataFrame( data = dados ).hist( bins = bins , density =1)
plt.plot( x2 , y2 , linewidth =3 , label = ' gaussiana ')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

## __3.7 Amostragens de distribuições discretas. Distribuição binomial__

```python
from scipy.stats import binom

amostra =[]
x =[]
n =60000
p =0.4
k =20
k1 = np.arange (0 ,20)
y1 = binom.pmf ( k1 ,20 , p )
for i in range (0 , n ) :
    binomial =0
    for j in range (0 , k ):
        r = random ()
        if r <= p :
            binomial +=1
    amostra.append ( binomial )
    x.append( i )
plt.plot (x , amostra , '-o')
```

```python
dados ={ 'y': amostra }
bins = np.arange (0, 20)
df = pd.DataFrame ( data = dados )
df = pd.cut ( df [ 'y'] , bins = bins )
df = df.groupby ( df ).size ()
df = df/df.sum ()
df.plot( kind = 'bar')
plt.plot ( k1 -1 , y1 , '-o' , color = 'orange' , linewidth =3)
```

## __3.8 Amostragens de distribuições discretas. Distribuição de Poisson__

```python
from scipy.stats import poisson

amostra =[]
x =[]
n =20000
lamb =5
L = math.exp ( - lamb )
k1 = np.arange (0 ,20)
y1 = poisson.pmf ( k1 , lamb )
for i in range (0, n) :
    p =1
    k =0
    while p >= L :
        r = random ()
        p = p * r
        k +=1
    amostra.append (k -1)
    x.append ( i )
plt.plot (x , amostra , '-o')
```

```python
dados ={ 'y ': amostra }
bins = np.arange (0 ,20)
df = pd.DataFrame ( data = dados )
df = pd.cut ( df [ 'y '] , bins = bins )
df = df.groupby ( df ).size ()
df = df/df.sum ()
df.plot ( kind = 'bar')
plt.plot ( k1 -1 , y1 , '-o' , color = 'orange' , linewidth =3)
plt.xlabel ( ' probabilidade de ocorrencia ')
plt.ylabel ( ' Poisson ')
```
