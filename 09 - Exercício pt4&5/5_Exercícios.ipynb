__1.__ Crie um gerador de números aleatórios em python, e gere uma sequência de 20 algarismos entre 1 e 6. Lance um dado 20 vezes e guarde os números. Compare as duas distribuições encontradas.

__2.__ Determine o valor de π utilizando um método de rejeição.

```python
from random import random
import math
import numpy as np

err =[]
pi3 =[]

for N in range (1, 3000, 1):
    area = 0
    pi2 = 0
    for i in range(N):
        x = -1+2*random() # x aleatório
        y = -1+2*random() # y aleatório
        r = x**2+y**2 # r aleatório
        
        if r <=1:
            area = area + 1
            
    pi2 = 4/N * area
    pi3.append( pi2 )
    err.append( np.sqrt ( np.var ( pi3 ) ) )
print ( ' valor de pi = ' , math.pi )
print ( ' resultado = ' , pi2 )
print ( ' erro = ' , np.sqrt ( np.var ( pi3 ) ) )
```

#### __5.__ Considere a seguinte distribuição de probabilidades (Lorentziana com centro em μ e meia largura a meia altura de γ):

$$ p(x) = K · \frac{1}{(\frac{x−μ}{γ})²+ 1}            −∞<x<∞ $$

__a.__ Normalize a distribuição (i.e. determine o valor da constante K que normaliza a distribuição).

mu = 0

p = K * (1/(( (x-mu) /gama)**2+1))

integrado
K*gama*(np.arctan((x-mu)/gama))    #tg^-1(-inf)=-pi/2  ||   tg^-1(inf)=pi/2
K*gama*np.pi = 1
K = 1/(gama*np.pi)

__b__ Utilizando o método da __transformada inversa__, escreva um código que permita amostrar desta distribuição, com μ=0,5 e γ=0,2

```python
amostra =[]
x =[]
mu = 0.5
gama = 0.2
K = 1/(gama*np.pi)
x2 = np.arange (-5, 5 ,0.01)
y2 = K*(1/(((x2-mu)/gama)**2+1))
n = 50000
for i in range (1, n):
    r = random()
    y = (gama*(np.tan(r/(K*gama)))) + mu ## Y é cdf = F^-1(x) entre 0 e 1    
    x.append(i)
    amostra.append(y)
dados ={ 'y': amostra }
bins = np.arange (-5, 5, .01)
df = pd.DataFrame ( data = dados ).hist ( bins = bins , density =1)
plt.plot ( x2 , y2 , linewidth =3 , label = ' Loretziana ')
plt.legend ()
plt.xlabel ('x')
plt.ylabel ('f(x)')
```

__6.__ O movimento browniano é um processo de Wiener, ou seja, um processo estocástico contínuo no tempo. Seja X(t) uma variável aleatória que depende de t ∈ [0, T], tal que:

X(0) = 0, 3, condição inicial
<p>X(t) ≈ X(s) + √t − s · N (0, 1).

Onde 0 < si < ti < .... < T , e N (μ, σ2), e é uma distribuição gaussiana padrão.

__a.__ Escreva um código onde utiliza a transformada de __Box-Müller__ (método de Marsaglia) para amostrar números aleatórios de uma gaussiana padrão.

```python
amostra1 =[]
amostra2 =[]
x =[]
mu = 0
sigma = 1
x2 = np.arange(-5, 5, 0.01)
y2 = 1/(np.sqrt(2*np.pi)*sigma)*np.exp(-0.5*(((x2-mu)/sigma))**2) #f(x)
n = 200000
for i in range(0, n):
    r1 = -1 + 2*random() # [-1 e 1]
    r2 = -1 + 2*random()
    s = r1**2 + r2**2
    if s!=0 and s<=1:
        s1 = -2* math.log(s)/s
        z1 = r1 * (s1**0.5) # X
        z2 = r2 * (s1**0.5) # Y
        amostra1.append(z1)
        amostra2.append(z2)
        x.append( i )
# plt.plot (x , amostra1 , '-o')
dados ={'y':amostra1}
bins = np.arange( -5 ,5 ,.01)
df = pd.DataFrame( data = dados ).hist( bins = bins , density =1)
plt.plot( x2 , y2 , linewidth =3 , label = ' gaussiana ')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```
