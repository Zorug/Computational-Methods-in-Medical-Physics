Vamos amostrar uma função exponencial $f(x)=e^{-x}$ a partir de uma distribuição uniforme $g(x)=c\cdot x$

```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
from random import random
from scipy.stats import norm
amostra=[]
amostra2=[]
x=[]
n=800000000
j=0
for i in range(1,n):
    x1=10*random()
    f=math.exp(-x1)
    y=1000*random()
    if y <= f:
        j+=1
        amostra.append(x1)
        x.append(j)
#plt.plot(x,amostra,'-o')
dados={'y':amostra}
bins=np.arange(0,10,.01)
df=pd.DataFrame(data=dados).hist(bins=bins)  
```

Para uma gaussiana numa caixa:

```python
%matplotlib inline
import os
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
from random import random
from scipy.stats import norm
amostra=[]
amostra2=[]
x=[]
n=80000
j=0
for i in range(1,n):
    x1=random()
    y=2*random()
    u=norm.pdf(x1,0.5,0.2)
    if y<=u:
        amostra.append(x1)
dados={'y':amostra}
bins=np.arange(0,1,.01)
df=pd.DataFrame(data=dados).hist(bins=bins)  
```
