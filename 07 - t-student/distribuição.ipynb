```python
%matplotlib inline
import os
import numpy as np
import math
from scipy.stats import t # modulo que possui a funcao t - student
import matplotlib.pyplot as plt

nu = np.arange(1 ,10 ,2)
def f(x , nu):
    return t.pdf(x , nu)
x = np.arange(-10 ,10 ,.1)

for nuval in nu :
    plt.plot(x , f(x , nuval) , '-' , label = r"$\nu$" f"={ nuval}")
    plt.legend()
plt.xlabel('x')
plt.ylabel('Probability density function')
plt.savefig('tdist1.png')
```
